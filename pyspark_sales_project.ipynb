```python
from pyspark.sql.types import (StructType, StructField, StringType, DateType, DoubleType, IntegerType)
```


**StructField(name, dataType, nullable)**
- StructField("order_id", IntegerType(), True)


```python
sale_schema = StructType([
    StructField("Product_id", IntegerType(), True),
    StructField("Customer_id", StringType(), True),
    StructField("Order_date", DateType(), True),
    StructField("Location", StringType(), True),
    StructField("Source_order", StringType(), True)
])
```

```python
menu_schema = StructType([
    StructField("Product_id", IntegerType(), True),
    StructField("Product_name", StringType(), True),
    StructField("Price", StringType(), True),
])
```

```python
sales_df_schema = spark.read.csv("/Workspace/Users/lotad92404@1200b.com/sales_project/sales.csv", schema=sale_schema) 
```

```python
menu_df_schema = spark.read.csv("/Workspace/Users/lotad92404@1200b.com/sales_project/menu.csv", schema=menu_schema)
```

```python
display(sales_df_schema)
```

```python
display(menu_df_schema)
```

> -  Importing lib for the date columns manupulation


```python
from pyspark.sql.functions import month, year, quarter
```

```python
sales_df_schema = sales_df_schema.withColumn("year", year(sales_df_schema.Order_date))
display(sales_df_schema)
```

```python
sales_df_schema = sales_df_schema.drop("order_month", "order_months", "order_year")
display(sales_df_schema)
```

```python
sales_df_schema = sales_df_schema.withColumn("order_months", month(sales_df_schema.Order_date))
sales_df_schema = sales_df_schema.withColumn("quarter", quarter(sales_df_schema.Order_date))
display(sales_df_schema)
```

### Join both df 
> base on the requirements

-  1. Total Amount Spent by the each customer
-  2. Also importing the Sum function for performing the sum while doing aggregation n col

```python
from pyspark.sql.functions import sum, col

total_amount_spent = (sales_df_schema
                      .join(menu_df_schema, 'Product_id')
                      .groupBy("Customer_id")
                      .agg(sum("Price").alias("total_amount_spent"))
)
```

```python
# The sum aggregation may show null if all Price values for a Customer_id are null or if there is an integer overflow.
display(total_amount_spent)
```

Total amount spent by each food category



```python
total_food_amount_spent = (sales_df_schema
                      .join(menu_df_schema, 'Product_id')
                      .groupBy("Product_name")
                      .agg(sum("Price").alias("total_food_amount_spent"))
                      .orderBy(col("total_food_amount_spent").desc())
)
```

```python
display(total_food_amount_spent)
```

###  total amount of the sale in the each month

```python
df_1 = (sales_df_schema.join(menu_df_schema, 'Product_id')
                      .groupBy("order_months")
                      .agg(sum("Price").alias("total_amount_spent_each_month"))
                      .orderBy("order_months")
)
display(df_1)
```

- yearly sales 

```python
total_spent_year = (sales_df_schema
                      .join(menu_df_schema, 'Product_id')
                      .groupBy("year")
                      .agg(sum("Price").alias("total_spent_year"))
                      .orderBy(col("total_spent_year").desc())
)
display(total_spent_year)
```

- total amount spent in quarterly year

```python
total_amount_spent_quarterly = (
                        sales_df_schema
                      .join(menu_df_schema, 'Product_id')
                      .groupBy("quarter")
                      .agg(sum("Price").alias("total_amount_spent_quarterly"))
                      .orderBy(col("total_amount_spent_quarterly").desc())
)
display(total_amount_spent_quarterly)
```

- how many time each product purchase 

```python
from pyspark.sql.functions import count

most_df = (sales_df_schema
           .join(menu_df_schema, "Product_id")
           .groupBy("Product_id", "product_name")
           .agg(count("Product_id").alias("No_Times_Product_Purchase"))
           .orderBy(col("No_Times_Product_Purchase").desc())
)

display(most_df)
```

- Top 5 orderer item

```python
most_5_df = (sales_df_schema
           .join(menu_df_schema, "Product_id")
           .groupBy("Product_id", "product_name")
           .agg(count("Product_id").alias("No_Times_Product_Purchase"))
           .orderBy(col("No_Times_Product_Purchase").desc()).limit(5)
           .drop("product_id")
)

display(most_5_df)
```

- Top Sold Item 

```python
top_sold_df = (sales_df_schema
           .join(menu_df_schema, "Product_id")
           .groupBy("Product_id", "product_name")
           .agg(count("Product_id").alias("No_Times_Product_Purchase"))
           .orderBy(col("No_Times_Product_Purchase").desc()).limit(1)
           .drop("Product_id")
)

display(top_sold_df)
```

- No Of time customer visiting restaurant

```python
from pyspark.sql.functions import count

df_visiting = (sales_df_schema.filter(sales_df_schema.Source_order == "Restaurant")
      .groupBy("Customer_id")
      .agg(count("Source_order").alias("No_of_Time_Customer_Visited"))
)

display(df_visiting)
```

- total sale by each country 

```python
total_sale_per_country = (sales_df_schema
                      .join(menu_df_schema, 'Product_id')
                      .groupBy("Location")
                      .agg(sum("Price").alias("total_sale_per_country"))
)

display(total_sale_per_country)
```

- total sale by order Source

```python
total_sale_By_Source = (sales_df_schema
                      .join(menu_df_schema, 'Product_id')
                      .groupBy("Source_order")
                      .agg(sum("Price").alias("total_sale_By_Source = (sales_df_schema"))
)
display(total_sale_By_Source)
```
